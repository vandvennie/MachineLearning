```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
# 1. 读取文件内容并进行数据检查/探索性数据分析
# 读取数据集
columns = ['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings']
data = pd.read_csv('abalone.data', names=columns)


# 先处理文本列（提前到数据探索之前）
data = pd.get_dummies(data, columns=['Sex'])  # 独热编码处理文本列

# 1. 查看数据集的基本信息
print('数据基本信息：')
data.info()

# 查看数据集行数和列数
rows, columns = data.shape

# 查看缺失值
missing_values = data.isnull().sum()
print('数据缺失值信息：')
print(missing_values)

# 绘制各特征的直方图
data.hist(bins=50, figsize=(20, 15))
plt.show()

# 查看特征之间的相关性
correlation = data.corr()
print('特征相关性：')
print(correlation)

# 绘制相关性热力图
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 2. 进行适当的数据预处理

# 分离特征和目标变量
X = data.drop('Rings', axis=1)
y = data['Rings']

# 85/15 分割
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42
)


```

```python
# 3. 进行 85/15 随机分割以形成训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
# 4. 使用默认超参数训练决策树回归器
dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(X_train, y_train)


# 可视化决策树（展示前3层）
plt.figure(figsize=(20, 10))
plot_tree(
    dt_regressor, 
    max_depth=3, 
    feature_names=X.columns, 
    filled=True, 
    proportion=True
)
plt.title("Decision Tree Structure (Default Hyperparameters)")
plt.show()
```

```python
# 5. 选择一个性能指标并检查模型在训练和测试数据上的性能
train_pred = dt_regressor.predict(X_train)
test_pred = dt_regressor.predict(X_test)

train_mse = mean_squared_error(y_train, train_pred)
test_mse = mean_squared_error(y_test, test_pred)

print(f'Training MSE: {train_mse}')
print(f'Test MSE: {test_mse}')

```

```python
# 6. 提供一个包含模型预测和测试集“真实”值的图表
plt.figure(figsize=(10, 6))
plt.scatter(y_test, test_pred, alpha=0.6)
plt.xlabel("True Values (Rings)")
plt.ylabel("Predicted Values")
plt.title("Decision Tree: True vs Predicted Values (Test Set)")
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)  # 对角线参考线
plt.show()
```

```python
# 7. 使用 5 折交叉验证，尝试不同的超参数并观察它们对模型性能的影响
param_grid = {
    'max_depth': [2, 5, 10, 15, 20],
    'min_samples_split': [2, 5, 10, 15, 20]
}

grid_search = GridSearchCV(DecisionTreeRegressor(random_state=42), param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print('Best hyperparameters:', best_params)

# 使用最优超参数重新训练模型
best_dt_regressor = grid_search.best_estimator_
best_train_pred = best_dt_regressor.predict(X_train)
best_test_pred = best_dt_regressor.predict(X_test)

best_train_mse = mean_squared_error(y_train, best_train_pred)
best_test_mse = mean_squared_error(y_test, best_test_pred)

print(f'Training MSE (Best Hyperparameters): {best_train_mse}')
print(f'Test MSE (Best Hyperparameters): {best_test_mse}')
```

```python
# 8. 为测试集提供一个包含两个模型（默认超参数和最优超参数）的预测和“真实”值的图表
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(y_test, test_pred)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True Values vs Predictions (Default Hyperparameters)')

plt.subplot(1, 2, 2)
plt.scatter(y_test, best_test_pred)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True Values vs Predictions (Best Hyperparameters)')

plt.tight_layout()
plt.show()
```

```python
# 9. 可视化微调后的决策树模型以了解它如何进行预测
plt.figure(figsize=(20, 10))
plot_tree(best_dt_regressor, max_depth=3, feature_names=X.columns, filled=True)
plt.show()
```

```python
# 10. 使用 SVM 回归器完成相同的任务，并使用 5 折交叉验证微调一些超参数
# 特征缩放
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

param_grid_svm = {
    'C': [0.1, 1, 10],
    'gamma': [0.01, 0.1, 1]
}

grid_search_svm = GridSearchCV(SVR(), param_grid_svm, cv=5, scoring='neg_mean_squared_error')
grid_search_svm.fit(X_train_scaled, y_train)

best_params_svm = grid_search_svm.best_params_
print('Best hyperparameters for SVM:', best_params_svm)

best_svm_regressor = grid_search_svm.best_estimator_
svm_train_pred = best_svm_regressor.predict(X_train_scaled)
svm_test_pred = best_svm_regressor.predict(X_test_scaled)

svm_train_mse = mean_squared_error(y_train, svm_train_pred)
svm_test_mse = mean_squared_error(y_test, svm_test_pred)

print(f'Training MSE (SVM): {svm_train_mse}')
print(f'Test MSE (SVM): {svm_test_mse}')

# 比较两个模型的性能
print('Decision Tree Test MSE:', best_test_mse)
print('SVM Test MSE:', svm_test_mse)

if best_test_mse < svm_test_mse:
    print('Decision Tree performed better on the test set.')
else:
    print('SVM performed better on the test set.')    
```
