```python
# 导入基础库
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.size'] = 12
```

```python
# 1. 加载训练集和测试集
df_train = pd.read_csv('training.csv')
df_test = pd.read_csv('testing.csv')
```

```python
# 2. 查看数据基本信息
print("Training set column headings:", df_train.columns)
print("First few lines of training set:")
print(df_train.head())
print("Info of training set:")
print(df_train.info())

print("\nTest set column headings:", df_test.columns)
print("First few lines of test set:")
print(df_test.head())
print("Info of test set:")
print(df_test.info())
```

```python
# 3. 数据清洗：删除无关列
columns_to_drop = [col for col in df_train.columns if col.startswith('pred_minus_obs')]
df_train = df_train.drop(columns=columns_to_drop)
df_test = df_test.drop(columns=columns_to_drop)
```

```python
# 4. 检查类别分布与数据平衡
class_counts_train = df_train['class'].value_counts()
class_counts_test = df_test['class'].value_counts()
print("\nClass counts in training set:", class_counts_train)
print("Class counts in test set:", class_counts_test)
is_imbalanced_train = (class_counts_train.max() / class_counts_train.min() > 10)
is_imbalanced_test = (class_counts_test.max() / class_counts_test.min() > 10)
print("Is training set imbalanced:", is_imbalanced_train)
print("Is test set imbalanced:", is_imbalanced_test)
```

```python
# 特征与标签提取
X_train = df_train[['b1', 'b2', 'b3', 'b4', 'b5', 'b6', 'b7', 'b8', 'b9']]
y_train = df_train['class']
X_test = df_test[['b1', 'b2', 'b3', 'b4', 'b5', 'b6', 'b7', 'b8', 'b9']]
y_test = df_test['class']
```

```python
# ====================== 任务1：特征缩放 ======================
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# ====================== 任务2-4：SVM处理 ======================
# 任务2：SVM网格搜索
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
param_grid_svm = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf', 'poly']}
svm = SVC()
grid_svm = GridSearchCV(svm, param_grid_svm, cv=3, scoring='accuracy')
grid_svm.fit(X_train_scaled, y_train)
```

```python
# 任务3：最佳超参数与验证性能
print("\nBest Hyperparameters for SVM:", grid_svm.best_params_)
for fold in range(3):
    print(f"{fold+1}Verification accuracy: {grid_svm.cv_results_['split{0}_test_score'.format(fold)][grid_svm.best_index_]}")
```

```python
# 任务4：训练评估
from sklearn.metrics import confusion_matrix
best_svm = grid_svm.best_estimator_
train_acc_scaled = best_svm.score(X_train_scaled, y_train)
test_acc_scaled = best_svm.score(X_test_scaled, y_test)
y_pred_scaled = best_svm.predict(X_test_scaled)
conf_mat_scaled = confusion_matrix(y_test, y_pred_scaled)
```

```python
# ====================== 任务5：无缩放SVM对比 ======================
grid_svm_no_scale = GridSearchCV(SVC(), param_grid_svm, cv=3, scoring='accuracy')
grid_svm_no_scale.fit(X_train, y_train)
best_svm_no_scale = grid_svm_no_scale.best_estimator_
test_acc_no_scale = best_svm_no_scale.score(X_test, y_test)
```

```python
# ====================== 任务6：特征缩放分析 ======================
print("\n Necessity analysis of feature scaling：")
print(f"SVM test accuracy after scaling: {test_acc_scaled:.4f}")
print(f"Unscaled SVM test accuracy: {test_acc_no_scale:.4f}")
print("Conclusion: Feature scaling improves SVM performance because SVM is sensitive to feature scale and parameter optimization is more effective after scaling.")
```

```python
# ====================== 任务7：逻辑回归处理 ======================

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

param_grid_lr = {'penalty': ['l2'], 'C': [0.1, 1, 10]}  # 若仍用lbfgs，penalty保留l2
lr = LogisticRegression(solver='lbfgs')  # 移除multi_class参数
grid_lr = GridSearchCV(lr, param_grid_lr, cv=3, scoring='accuracy')
grid_lr.fit(X_train_scaled, y_train)

best_lr = grid_lr.best_estimator_
test_acc_lr = best_lr.score(X_test_scaled, y_test)
```

```python
# ====================== 任务8：K-NN处理 ======================
from sklearn.neighbors import KNeighborsClassifier
param_grid_knn = {'n_neighbors': [3, 5, 7]}
knn = KNeighborsClassifier()
grid_knn = GridSearchCV(knn, param_grid_knn, cv=3, scoring='accuracy')
grid_knn.fit(X_train_scaled, y_train)

best_knn = grid_knn.best_estimator_
test_acc_knn = best_knn.score(X_test_scaled, y_test)
```

```python
# ====================== 任务9：结果总结 ======================
print("\nModel Performance Summary：")
print(f"SVM test accuracy: {test_acc_scaled:.4f}")
print(f"Logistic regression test accuracy: {test_acc_lr:.4f}")
print(f"K-NN test accuracy: {test_acc_knn:.4f}")
```

```python

import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="matplotlib")

# 可视化SVM混淆矩阵
plt.figure(figsize=(8, 6))
sns.heatmap(conf_mat_scaled, annot=True, fmt='d', cmap='Blues', 
            xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.xlabel('Predicted Class')
plt.ylabel('True Class')
plt.title('SVM Confusion Matrix')
plt.show()
```
